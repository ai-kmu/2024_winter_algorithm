```python
class MyHashMap:

    def __init__(self):
        self.list = [None] * 1000001


    def put(self, key, value):
        self.list[key] = value

    def get(self, key):
        value = self.list[key]
        if value != None:
            return value
        else:
            return -1

    def remove(self, key) :
        self.list[key] = None



hashmap = MyHashMap()
print(hashmap.put(1, 1))
print(hashmap.put(2, 2))
print(hashmap.get(1))
print(hashmap.get(3))
print(hashmap.put(2, 1))
print(hashmap.get(2))
print(hashmap.remove(2))
print(hashmap.get(2))
```
